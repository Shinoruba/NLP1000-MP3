- 0 = **Match**
- 1 = **Deletion**
- 2 = **Insertion**
- 3 = **Substitution**

```
def minimum_edit_distance(source, target):
    # your code here
    n, m = len(source), len(target)
    dp = [[0] * (m + 1) for _ in range(n + 1)]
    backtrace = [[0] * (m + 1) for _ in range(n + 1)]

    for i in range(n + 1):
        dp[i][0] = i
        backtrace[i][0] = 1  # deletion
    for j in range(m + 1):
        dp[0][j] = j
        backtrace[0][j] = 2  # insertion

    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if source[i-1] == target[j-1]:
                dp[i][j] = dp[i-1][j-1]
                backtrace[i][j] = 0  # match
            else:
                deletion = dp[i-1][j] + 1
                insertion = dp[i][j-1] + 1
                substitution = dp[i-1][j-1] + 2 # mismatch

                dp[i][j] = min(deletion, insertion, substitution)

                if dp[i][j] == substitution:
                    backtrace[i][j] = 3  # substitution
                elif dp[i][j] == insertion:
                    backtrace[i][j] = 2  # insertion
                else:
                    backtrace[i][j] = 1  # deletion

# =======================================================================
    align_source = []
    align_target = []
    align_middle = []
    i, j = n, m

    while i > 0 or j > 0:
        if backtrace[i][j] == 0:    # Match
            align_source.append(source[i-1])
            align_target.append(target[j-1])
            align_middle.append(' ')
            i -= 1
            j -= 1
        elif backtrace[i][j] == 1:  # Deletion
            align_source.append(source[i-1])
            align_target.append('-')
            align_middle.append('D')
            i -= 1
        elif backtrace[i][j] == 2:  # Insertion
            align_source.append('-')
            align_target.append(target[j-1])
            align_middle.append('I')
            j -= 1
        else:                       # Substitution
            align_source.append(source[i-1])
            align_target.append(target[j-1])
            align_middle.append('M')
            i -= 1
            j -= 1

    # reverse alignments
    align_source = ''.join(reversed(align_source))
    align_target = ''.join(reversed(align_target))
    align_middle = ''.join(reversed(align_middle))

    return dp[n][m], align_source, align_middle, align_target


# Note: you're not forced to use this template, so you can implement multiple
#       functions or rename functions as you see fit.
# Thank you, this note is noted :)
```
